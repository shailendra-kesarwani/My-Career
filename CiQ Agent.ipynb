```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings, StorageContext, load_index_from_storage
from llama_index.llms.ollama import Ollama
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
import os

# --- Configuration ---
# Set the LLM and Embedding models
Settings.llm = Ollama(model="llama3.2:1b", request_timeout=60.0) # You can use other Llama 3.2 models like 3B, 11B, etc.
Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-large-en-v1.5")
Settings.chunk_size = 512
Settings.chunk_overlap = 50

# --- Create Knowledgebase and Indexing (Creating a Vector Store) ------
# Place your PDF files in a 'docs' directory

# This step converts your documents into embeddings and stores them
PERSIST_DIR = "./knowledge-storage"
if not os.path.exists(PERSIST_DIR):
    # Load dental knowledge documents and create index
    documents = SimpleDirectoryReader("docs").load_data()
    index = VectorStoreIndex.from_documents(documents)
    index.storage_context.persist(persist_dir=PERSIST_DIR)
else:
    # Load existing dental knowledge index
    storage_context = StorageContext.from_defaults(persist_dir=PERSIST_DIR)
    index = load_index_from_storage(storage_context)

# --- Query Engine ---
# Create a query engine to interact with your data
query_engine = index.as_query_engine()

# --- Interaction (RAG in action) ---
question = "You are a CiQ officer whose job is to evaluate the DiQ Sanction screening system independently for a Fin-Tech organization in AML compliance space. You can analyse the CSV and create a Report based on a template HTML report."



# Example of asking another question
question += open('Ciq_prompts.txt', 'r').read()
print(f"Question: {question}")
# response = query_engine.query(question)
# print(f"Question: {question_2}")
# print(f"Response: {response_2}")
```
